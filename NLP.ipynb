```python
# NER
# Install once
# pip install spacy pandas
# python -m spacy download en_core_web_sm

import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm")

df = pd.read_csv(
    "data.csv",
    engine="python",
    on_bad_lines="skip"
)

# ðŸ‘‡ USE A REAL TEXT COLUMN
text = df["description"].dropna().astype(str).iloc[0]
# OR try: df["title"]

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
#Summarization
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd

# load dataset safely
df = pd.read_csv(
    "data.csv",
    engine="python",
    on_bad_lines="skip"
)

# use a real text column
text = df["description"].dropna().astype(str).iloc[0]

model_name = "sshleifer/distilbart-cnn-12-6"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# tokenize input
inputs = tokenizer(
    [text],
    max_length=1024,
    return_tensors="pt",
    truncation=True
)

# generate summary
summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4,
    max_length=50,
    min_length=25,
    early_stopping=True
)

# decode summary
summary_text = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print(summary_text)
```

```python
# Install once
# pip install nltk pandas

import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# SAFE CSV LOAD (handles broken quotes)
df = pd.read_csv(
    "data.csv",
    engine="python",        # important
    on_bad_lines="skip"     # skip corrupted rows
)

# Check columns once (optional)
print(df.columns)

# CHANGE this to your actual text column
text_column = df.columns[0]   # uses first column automatically

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Take one clean text
text = str(df[text_column].dropna().iloc[0])

# Sentiment score
score = sia.polarity_scores(text)
print(score)

# Final sentiment
if score['compound'] >= 0.05:
    print("Positive")
elif score['compound'] <= -0.05:
    print("Negative")
else:
    print("Neutral")
```
