```python
#TASK1:Perform any operations and write a program with two examples
#1)Area of a circle
def circle_area(r):
   area=3.14*r*r
   return area
print(circle_area(6))
#2)Perimeter of rectangle
def perimeter_of_rectangle(l,b):
   perimeter=2*(l+b)
   return perimeter
print(perimeter_of_rectangle(5,6))
```

```python
#TASK2:Program on object and class(real-life usecase)
class Person():
    def __init__(self,name,height,weight):
        self.name=name
        self.height=height
        self.weight=weight
        self.is_studying=True

    def __str__(self):
        study_status = "studying" if self.is_studying else "not studying"
        return f"{self.name} (Height: {self.height}, Weight: {self.weight}), {study_status}"

    #method:age of the person
    def age_no(self):
        return f"{self.name} is 7 years old"

    #method:study the person
    def study(self):
        if self.is_studying:
            self.is_studying=False
            return f"{self.name} is studying"
        else:
            return f"{self.name} is not studying"
    #method:height and weight of the person
    def height_weight(self):
        return f"{self.name} is {self.height} in height and {self.weight} in weight"

#creating objects from class
if __name__ =="__main__":
    p1=Person(name="shruthi",height=5.8,weight=60)
    p2=Person(name="sahithi",height=6.0,weight=65)
    print(p1)
    print(p2)
    print(p1.age_no())
    print(p2.study())
    print(p1.height_weight())
```
